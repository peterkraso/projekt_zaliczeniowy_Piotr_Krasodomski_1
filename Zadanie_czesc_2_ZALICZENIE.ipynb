```python
import pandas as pd

df = pd.read_csv("dane1.csv")
df.shape, df.columns
```

```python
### PODPUNKT 1
df_sorted = df.sort_values("favorite_count", ascending = False)                   # Sortowanie danych po liczbie polubien

top5_likes = df_sorted.head(5)                                                    # Wybranie TOP 5 tweetow

top5_likes[["favorite_count", "text", "tweet_url"]].reset_index(drop = True)      # Wyswietlenie podanych(sensownych) kolumn + uporzadkowanie
```

```python
### PODPUNKT 2
df_sorted = df.sort_values("retweet_count", ascending = False)                    # Sortowanie po liczbie retweetow

top5_retweets = df_sorted.head(5)                                                 # Wybranie TOP 5 retweetow

top5_retweets[["retweet_count", "text", "tweet_url"]].reset_index(drop = True)    # Wyswietlenie kolumn + uporzadkowanie
```

```python
### PODPUNKT 3
not_sensitive = df[df["possibly_sensitive"] == False]                                              # Filtrowanie tylko FALSE

not_sensitive[["possibly_sensitive", "text", "tweet_url"]].head(10).reset_index(drop = True)       # Wyswietlenie kolumn z uporzadkowaniem
```

```python
### PODPUNKT 4
df["user_created_at_dt"] = pd.to_datetime(                                        # zmiana user_created_at na date
    df["user_created_at"],
    errors = "coerce"
)

earliest_date = df["user_created_at_dt"].min()                                    # znaleznienie najwczesniej zalozonego konta
earliest_date


earliest_user_tweets = df[df["user_created_at_dt"] == earliest_date]              # wybranie tweeta danego uzytkownika

earliest_user_tweets[

                     ["user_created_at", "user_name", "text", "tweet_url"]        # wyswietlenie uporzadkowanych kolumn
].reset_index(drop = True)
```

```python
### PODPUNKT 5
max_followers = df["user_followers_count"].max()                                  # znalezienie maksymalniej liczby followow
max_followers

top_follow_user_tweets = df[df["user_followers_count"] == max_followers]          # wybranie tweetow danego uzytkownika

top_follow_user_tweets[
    ["user_followers_count", "user_name", "text", "tweet_url"]                    # wyswietlenie tweetow
].reset_index(drop = True)
```

```python
### PODPUNKT 6
verified_users = df[df["user_verified"] == True]                                  # filtrowanie zweryfikowanych userow

verified_users[
    ["user_verified", "user_name", "text", "tweet_url"]                           # wyswietlenie top 10 uzytkownikow
].head(10).reset_index(drop = True)
```

```python
###PODPUNKT 7
df["created_at_dt"] = pd.to_datetime(
    df["created_at"],                                                             # zamiana created_at na date
    errors = "coerce"
)

df["weekday_en"] = df["created_at_dt"].dt.day_name()                              # wyciagniecie dzien tygodnia

day_map_en_pl = {
    "Monday": "poniedzialek",
    "Tuesday": "wtorek",
    "Wednesday": "sroda",
    "Thursday": "czwartek",                                                       # zmiana na polskie nazwy
    "Friday": "piatek",
    "Saturday": "sobota",
    "Sunday": "niedziela",
}
df["created_at_day_pl"] = df["weekday_en"].map(day_map_en_pl)

day_conts = df["created_at_day_pl"].value_counts()
most_common_day = day_conts.idxmax()                                              # policzenie i wybranie najczestszego dnia

day_conts, most_common_day

### dzien w ktorym najczesciej tweetowano to: SRODA 1657 tweetow
```
